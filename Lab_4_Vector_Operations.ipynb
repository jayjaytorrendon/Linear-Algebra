```
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
print("Task 1: Modulus Operation[1]")
print("Task 2: Inner Product[2]")
print("Task 3: Resultant Vector[3]")
select=input("Select Task: ")
#Task 1
if select=='1':
  a = np.array([3,8,6,7])
  b = np.array([6,4,8,9])
  c = np.array([5,4,3,7])
  d = np.array([2,1,0,4])
  e = np.array([6,9,4,5])
  f = np.array([3,7,3,3])
  print("A: [a]",a)
  print("B: [b]",b)
  print("C: [c]",c)
  print("D: [d]",d)
  print("E: [e]",e)
  print("F: [f]",f)
  aselect=input("Select Vector: ")
  if aselect=='a':
    print("Vector A: ",a)
    modula = a*a
    suma= 0
    for i in range(0,len(modula)):
      suma = suma + modula[i]
    fina=suma**.5
    print("Modulus of vector A: ",fina)
  elif aselect=='b':
    print("Vector B: ",b)
    modulb = b*b
    sumb= 0
    for i in range(0,len(modulb)):
      sumb = sumb + modulb[i]
    finb=sumb**.5
    print("Modulus of vector B: ",finb)
  elif aselect=='c':
    print("Vector C: ",c)
    modulc = c*c
    sumc= 0
    for i in range(0,len(modulc)):
      sumc = sumc + modulc[i]
    finc=sumc**.5
    print("Modulus of vector C: ",finc)
  elif aselect=='d':
    print("Vector D: ",d)
    moduld = d*d
    sumd= 0
    for i in range(0,len(moduld)):
      sumd = sumd + moduld[i]
    find=sumd**.5
    print("Modulus of vector D: ",find)
  elif aselect=='e':
    print("Vector E: ",e)
    module = e*e
    sume= 0
    for i in range(0,len(module)):
      sume = sume + module[i]
    fine=sume**.5
    print("Modulus of vector E: ",fine)
  elif aselect=='f':
    print("Vector F: ",f)
    modulf = f*f
    sumf= 0
    for i in range(0,len(modulf)):
      sumf = sumf + modulf[i]
    finf=sumf**.5
    print("Modulus of vector F: ",finf)
  else:
    print("Invalid Selection")
#Task 2
elif select=='2':
  a2=np.array([6,9,8,5,7])
  b2=np.array([8,10,7,3,4])
  c2=np.array([9,10,5,3,2])
  d2=np.array([7,4,3,2,9])
  e2=np.array([9,10,6,5,4])
  print("A: ",a2)
  print("B: ",b2)
  print("C: ",c2)
  print("D: ",d2)
  print("E: ",e2)
  print("=========")
  print("Select Inner Product Combination")
  print("a. A \xb7 B")
  print("b. B \xb7 C \xb7 D")
  print("c. D \xb7 E \xb7 A \xb7 B")
  print("d. A \xb7 B \xb7 C \xb7 D \xb7 E")
  bselect=input("Select: ")
  if bselect=='a':
    print("{} \xb7 {}".format(a2,b2))
    ipa=a2*b2
    sa=0
    for i in range(0,len(ipa)):
      sa = sa + ipa[i]
    print("Inner Product of {} and {}: {}".format(a2,b2,sa))
  elif bselect=='b':
    print("{} \xb7 {} \xb7 {}".format(b2,c2,d2))
    ipb=b2*c2*d2
    sb=0
    for i in range(0,len(ipb)):
      sb = sb + ipb[i]
    print("Inner Product of {}, {} and {}: {}".format(b2,c2,d2,sb))
  elif bselect=='c':
    print("{} \xb7 {} \xb7 {} \xb7 {}".format(d2,e2,a2,b2))
    ipc=d2*e2*a2*b2
    sc=0
    for i in range(0,len(ipc)):
      sc = sc + ipc[i]
    print("Inner Product of {}, {}, {} and {}: {}".format(d2,e2,a2,b2,sc))
  elif bselect=='d':
    print("{} \xb7 {} \xb7 {} \xb7 {} \xb7 {}".format(a2,b2,c2,d2,e2))
    ipd=a2*b2*c2*d2*e2
    sd=0
    for i in range(0,len(ipd)):
      sd = sd + ipd[i]
    print("Inner Product of {}, {}, {}, {} and {}: {}".format(a2,b2,c2,d2,e2,sd))
  else:
    print("Invalid Selection")
#Task 3
elif select=='3':
  a3 = np.array([-0.4,4.3,-0.6])
  b3 = np.array([-0.20,0.21,0])
  c3 = np.array([-0.2,2.1,-1.5])
  print("A: ",a3)
  print("B: ",b3)
  print("C: ",c3)
  d=a3+b3+c3
  e=np.linalg.norm(d) 
  print("Magnitude: ",e)
  print("Resultant: ",d)
  fig = plt.figure(figsize=(8,8))
  ax = fig.gca(projection='3d')
  ax.set_xlim([0, -1.5])
  ax.set_ylim([0, 7])
  ax.set_zlim([0, -2])
  ax.quiver(0, 0, 0, a3[0], a3[1], a3[2], arrow_length_ratio=.05, label='A',colors='red')
  ax.quiver(0, 0, 0, b3[0], b3[1], b3[2], arrow_length_ratio=.05, label='B',colors='blue')
  ax.quiver(0, 0, 0, c3[0], c3[1], c3[2], arrow_length_ratio=.05, label='C',colors='green')
  ax.quiver(0, 0, 0, d[0], d[1], d[2], arrow_length_ratio=.05,label='Resultant', colors='gold')
  ax.set_xlabel('x-axis')
  ax.set_ylabel('y-axis')
  ax.set_zlabel('z-axis')
  plt.legend()
  plt.show()
else:
  print("Invalid Selection")

```
