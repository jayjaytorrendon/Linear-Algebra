```
import numpy as np
import matplotlib.pyplot as plt
print("Vector Operations and Graph")
print("============")
v1 = np.array([5,5])
v2 = np.array([3,4])
v3 = np.array([2,4])
v4 = np.array([3,2])
v5 = np.array([1,6])
print("Vector A is",v1)
print("Vector B is",v2)
print("Vector C is",v3)
print("Vector D is",v4)
print("Vector E is",v5)
print("============")
print("Operations:\
\nAddition and Subtraction[1]\
\nMultiplication and Division[2]\
\nSquaring and Square Root[3]\
\nSummation[4]")
op=input("Select Operation: ")
if op=='1':
  print("Select Vector Combinations:\
  \n\ta. A + B + C\
  \n\tb. B + D - E\
  \n\tc. D - B - C\
  \n\td. A - E + D")
  achoice = input("Enter choice: ")
  if achoice =='a':
    r= v1+v2+v3
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, v2[0], v2[1],angles='xy', scale_units='xy',scale=1,label='B', color='green')
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='b':
    r= v2+v4-v5
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v2[0], v2[1], angles='xy', scale_units='xy',scale=1,label='B', color='red')
    plt.quiver(0,0, v4[0], v4[1],angles='xy', scale_units='xy',scale=1,label='D', color='green')
    plt.quiver(0,0, v5[0], v5[1],angles='xy', scale_units='xy',scale=1,label='E', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='c':
    r= v4-v2-v3
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v4[0], v4[1], angles='xy', scale_units='xy',scale=1,label='D', color='red')
    plt.quiver(0,0, v2[0], v2[1],angles='xy', scale_units='xy',scale=1,label='B', color='green')
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='d':
    r= v1-v5+v4
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, v5[0], v5[1],angles='xy', scale_units='xy',scale=1,label='E', color='green')
    plt.quiver(0,0, v4[0], v4[1],angles='xy', scale_units='xy',scale=1,label='D', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
elif op=='2':
  print("Select Vector Combinations:\
  \n\ta. A * B\
  \n\tb. B * E / C \
  \n\tc. D / B / C\
  \n\td. A * E * D")
  achoice = input("Enter choice: ")
  if achoice =='a':
    r= v1*v2
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, v2[0], v2[1],angles='xy', scale_units='xy',scale=1,label='B', color='green')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='b':
    r= v2*v5/v3
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v2[0], v2[1], angles='xy', scale_units='xy',scale=1,label='B', color='red')
    plt.quiver(0,0, v5[0], v5[1],angles='xy', scale_units='xy',scale=1,label='E', color='green')
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='black')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='c':
    r= v4/v2/v3
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v4[0], v1[1], angles='xy', scale_units='xy',scale=1,label='D', color='red')
    plt.quiver(0,0, v2[0], v2[1],angles='xy', scale_units='xy',scale=1,label='B', color='green')
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='black')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='d':
    r= v1*v5*v4
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-70, 70)
    plt.ylim(-70, 70)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, v5[0], v5[1],angles='xy', scale_units='xy',scale=1,label='E', color='green')
    plt.quiver(0,0, v4[0], v4[1],angles='xy', scale_units='xy',scale=1,label='D', color='gold')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='black')
    plt.grid()
    plt.legend()
    plt.show()
elif op=='3':
  print("Select Vector Combinations:\
  \n\ta.  a^2 \
  \n\tb. sqr root of (a^2+b^2+c^2) ")
  achoice = input("Enter choice: ")
  if achoice =='a':
    r= v1**2
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-30, 30)
    plt.ylim(-30, 30)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='b':
    r = np.sqrt(np.sum(a**2+b**2+c**2))
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-30, 30)
    plt.ylim(-30, 30)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, v2[0], v2[1],angles='xy', scale_units='xy',scale=1,label='B', color='green')
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.grid()
    plt.legend()
    plt.show()
elif op=='4':
  print("Select Vector Combinations:\
  \n\ta. Summation of Vectors 1 to Vector 5 \
  \n\tb. Summation of Vectors 3 to 5")
  achoice = input("Enter choice: ")
  if achoice =='a':
    r= v1 + v2 + v3 + v4 + v5
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-29, 29)
    plt.ylim(-29, 29)
    plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy',scale=1,label='A', color='red')
    plt.quiver(0,0, v2[0], v2[1],angles='xy', scale_units='xy',scale=1,label='B', color='green')
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.quiver(0,0, v4[0], v4[1],angles='xy', scale_units='xy',scale=1,label='D', color='blue')
    plt.quiver(0,0, v5[0], v5[1],angles='xy', scale_units='xy',scale=1,label='E', color='yellow')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()
  elif achoice =='b':
    r=v3 + v4 + v5
    print("Result is: ",r)
    plt.title("Resultant Vector\nMagnitude: {}".format(r))
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.quiver(0,0, v3[0], v3[1],angles='xy', scale_units='xy',scale=1,label='C', color='gold')
    plt.quiver(0,0, v4[0], v4[1],angles='xy', scale_units='xy',scale=1,label='D', color='blue')
    plt.quiver(0,0, v5[0], v5[1],angles='xy', scale_units='xy',scale=1,label='E', color='yellow')
    plt.quiver(0,0, r[0], r[1],angles='xy', scale_units='xy',scale=1, label='R',color='aqua')
    plt.grid()
    plt.legend()
    plt.show()




```
